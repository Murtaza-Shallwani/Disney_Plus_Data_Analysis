```python
inline_rc = dict(mpl.rcParams)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_theme(style='white', palette='Set3')
sns.countplot(x = 'type', data = data);
```

```python
sns.set_theme(style='white')
sns.countplot(x = 'rating', data = data, palette = 'viridis')
```
